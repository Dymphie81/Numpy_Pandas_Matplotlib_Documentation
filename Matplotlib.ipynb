```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

```

```python
#Draw a line in a diagram from position (0,0) to position (5,250):
xpoints = np.array([0, 5])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# plotting without line
# Draw two points in the diagram, one at position (0, 2) and one in position (9, 12)

xpoints = np.array([0, 9])
ypoints = np.array([2, 12])
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
# To plot multiple points

x_points = np.array([1,4,6,8])
y_points = np.array([10,15,20,25])

plt.plot(x_points, y_points, 'o')
plt.show()
```

```python
# plotting without x-points

ypoints = np.array([5,10,1,12,2,7,15,6])
plt.plot(ypoints)
plt.show()

# mark each point with a circle
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
# To create a scatter plot

x = np.array([10,3,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([95,86,85,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
# Creating bar charts

x = np.array(["2000", "2001", "2002", "2003"])
y = np.array([50, 70, 30, 100])
plt.xlabel("Year")
plt.ylabel("Percentage of Maternal Mortality Rate")
plt.bar(x,y)
plt.show()

# shows the maternal mortality rate from the year 2000 to  year 2003

```
